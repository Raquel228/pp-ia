**c. Obtenga la media, mediana, moda y geométrica (el codigo se halla en incisoC.py); explique la diferencia de los resultados y cuál de ellas se puede utilizar en un artículo científico**

- **Media:** Es el promedio aritmético de los valores en la columna. Representa el valor central de un conjunto de datos y es sensible a los valores extremos. Para este caso, la media para cada columna indica el valor promedio de esa característica en todas las reservas de aerolíneas.
- **Mediana:** Es el valor que se encuentra en el centro de un conjunto de datos ordenado. Es menos sensible a los valores extremos que la media. En este caso, la mediana para cada columna indica el valor medio de esa característica, donde aproximadamente la mitad de las reservas tienen un valor menor y la otra mitad un valor mayor.
- **Moda:** Es el valor que aparece con mayor frecuencia en un conjunto de datos. Es útil para identificar el valor más común en una distribución. En este caso, la moda para cada columna indica el valor más frecuente de esa característica en las reservas de aerolíneas.
- **Media geométrica:** Es el promedio de los logaritmos de los valores en un conjunto de datos, útil para manejar datos positivos que varían exponencialmente, este estadístico es útil cuando se manejan datos con crecimiento exponencial, como tasas de crecimiento. Para este caso, la media geométrica para cada columna indica el valor promedio de esa característica, teniendo en cuenta su crecimiento exponencial.

En artículos científicos, el estadístico más comúnmente utilizado es la media aritmética. Esto se debe a que proporciona una medida central representativa de un conjunto de datos, lo que permite a los investigadores resumir y comparar valores. En el contexto del conjunto de datos de reservas de aerolíneas, la media aritmética sería útil para describir características como el número promedio de pasajeros por reserva, la duración promedio de la estancia o la hora promedio de salida de los vuelos. Utilizar la media aritmética en este contexto proporcionaría una descripción clara y concisa de las características de las reservas de aerolíneas, lo que facilitaría la interpretación y la comparación de los datos.


# Inciso D

La explicacion del comportamiento de los datos está en el documento pdf adjunto

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos CSV en un DataFrame de pandas
data = pd.read_csv("data.csv", encoding="latin1")

# Obtener las columnas del DataFrame
columnas = data.columns

# Iterar sobre cada columna y graficar
for col in columnas:
    # Verificar si la columna es categórica (es decir, tiene menos de 10 valores únicos)
    if data[col].nunique() < 10:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=col, data=data)
        plt.title(f'Distribución de {col}')
        plt.xlabel(col)
        plt.ylabel('Cantidad')
        plt.show()
    else:
        plt.figure(figsize=(8, 6))
        sns.histplot(data[col], bins=20, kde=True)
        plt.title(f'Distribución de {col}')
        plt.xlabel(col)
        plt.ylabel('Cantidad')
        plt.show()
```
